```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
results_resnet18 = pd.read_csv('/Users/nicenoize/Downloads/plantnet-main/results/results_Resnet18_30.csv')
results_resnet50 = pd.read_csv('/Users/nicenoize/Downloads/plantnet-main/results/results_Resnet50_30.csv')
results_resnet50_imbalanced = pd.read_csv('/Users/nicenoize/Downloads/results_Resnet50_30_Imbalanced.csv')
results_resnext = pd.read_csv('/Users/nicenoize/Downloads/plantnet-main/results/results_Resnext101_30.csv')
results_vit_20 = pd.read_csv('/Users/nicenoize/Downloads/plantnet-main/results/results_VisionTransformer_20.csv')
results_vit_40 = pd.read_csv('/Users/nicenoize/Downloads/plantnet-main/results/results_VisionTransformer_40.csv')
```

```python
results_resnet50_WRS = pd.read_csv('/Volumes/ExternalSD/Downloads/plantnet-main/results_weightedRandomSampler/results_Resnet50_30.csv')
```

```python
def plot(train_val, test_val, name, table, epochs):
    epochs = range(1,epochs+1)
    train = table[train_val]
    test = table[test_val]
    plt.plot(epochs, train, 'g', label= train_val)
    plt.plot(epochs, test, 'b', label= test_val)
    plt.title('Training and Validation' + name)
    plt.xlabel('Epochs')
    plt.ylabel(str(name))
    plt.legend()
    plt.savefig(name + '.png')
    plt.show()
```

```python
# Resnet50
plot('acc_train', 'acc_test', 'Accuracy Resnet50', results_resnet50_WRS, 30)
plot('loss_train', 'loss_test', 'Loss Resnet50', results_resnet50_WRS, 30)
plot('topk_acc_train', 'topk_acc_test', 'Top K Resnet50', results_resnet50_WRS, 30)
plot('f1_train', 'f1_test', 'F1 Score Resnet50', results_resnet50_WRS, 30)
```
